```python
#import opencv for computer vision staff
import cv2
#import matplotlib to visualize an image
from matplotlib import pyplot as plt
```

## Conneecting to laptop Camera

```python
# Connect to Capture device
cap = cv2.VideoCapture(0)
```

```python
#Get a frame from the capture device
ret, frame = cap.read()
```

```python
print(ret)
```

```python

frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame_rgb)
```

```python
# Releae capture back into the wild
cap.release()
```

```python
def take_photo():
    cap = cv2.VideoCapture(0)
    ret,frame= cap.read()
    cv2.imwrite('webcamphoto.jpg', frame)
    cap.release()
```

```python
take_photo()
```
